```python
#Most frecuent age and gender
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["Age"], bins=20, kde=True)
```

```python
bins = [18, 30, 40, 50, 60, 70, 80]  # define age group 
labels = ["18-30","31-40", "41-50", "51-60", "61-70", "71-80"] 

df["Age_Group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)
df
```

```python
sns.histplot(df["Age_Group"], bins=20)
```

```python
from plotnine import *

(
    ggplot(df) 
    + aes(x='Age_Group')  
    + geom_bar(fill="skyblue", color="black")  
    + theme_minimal() 
)
```

```python
#BMI Distrubition
sns.boxplot(y=df['BMI'])
plt.title("BMI Distribution")
plt.ylabel("BMI")
plt.show()
```

```python
#BMI by Gender 
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='BMI', data=df)
plt.title("BMI Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("BMI")
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_counts,labels=gender_counts.index, autopct='%1.1f%%') 
plt.title("Gender Distribution") 
plt.show()
```

```python
#Avg_BPM vs. Age

( ggplot(df) 
 +aes(x='Age',
      y='Avg_BPM')
 +geom_point()
 
)
```

```python
#Fat_Percentage vs. BMI

(
    ggplot(df)  # What data to use
    + aes(x = 'BMI', 
          y = 'Fat_Percentage'
      
         )  # What variable to use
    + geom_point()
    + geom_smooth(method = 'lowess')  #juntar puntos
    + geom_smooth(method = 'lm')# Geometric object to use for drawing
)
```

```python
sns.regplot(x=df["BMI"], y=df["Fat_Percentage"])
plt.title("BMI vs. Fat Percentage")
plt.xlabel("BMI")
plt.ylabel("Fat Percentage")
plt.show()
```

```python
#Exercise Habit Analysis

#Weekly Training Frequency vs. Gender

#Analyzing whether there are differences in training frequency between males and females


sns.barplot(x=df["Gender"], y=df["Workout_Frequency (days/week)"])
plt.title("Workout Frequency by Gender")
plt.ylabel("Workout Frequency (days/week)")
plt.show()
```

```python
# Training Duration vs. Calories Burned
# Chart Type: Scatter Plot + Regression Line
# Examine whether training duration is related to calories burned
(
    ggplot(df)  # What data to use
    + aes(x = 'Session_Duration (hours)', 
          y = 'Calories_Burned',
          color='Gender'
      
         )  # What variable to use
    + geom_point()
    + geom_smooth(method = 'lowess')  #juntar puntos
    + geom_smooth(method = 'lm')# Geometric object to use for drawing
)
```

```python
#Workout Type Analysis
#The impact of different types of workouts (Workout_Type) on calories burned and training duration.
#boxplot

(
    ggplot(df)
    + aes( x = 'Workout_Type',
         y = "Calories_Burned")
    + geom_boxplot()
    + coord_flip() 
    
)
```

```python
sns.boxplot(x=df["Workout_Type"], y=df["Calories_Burned"])
plt.title("Calories Burned by Workout Type")
plt.xticks(rotation=45)  
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 选取数值型数据
numerical_data = df.select_dtypes(include=['int64', 'float64'])

# 计算相关性矩阵
corr_matrix = numerical_data.corr()  #计算变量之间的皮尔逊相关系数（Pearson Correlation Coefficient）

# 绘制热力图
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# 设置标题
plt.title("Correlation Heatmap of Numerical Features")
plt.show()


#矩阵中的每个值表示两个变量之间的相关系数（correlation coefficient）
```
