```python
import random
import openpyxl
from getpass import getpass
import pandas as pd

#Menu principal
def Menu():
    print("ADIVINA EL NÚMERO")
    print("Menú principal: \n1.Partida modo solitario. \n2.Partida 2 Jugadores. \n3.Estadística. \n4.salir")
    option = int(input("Selecciona una opción: "))
    return option 

#Funcion Nivel de Dificultad
def dificultad():
    while True:
        print("1.Fácil (20 intentos)")
        print("2.Medio (12 intentos)")
        print("3.Difícil (5 intentos)") 
        dificultad = int(input("Selecciona el nivel de dificultad: "))
    
        if dificultad == 1:
            return 20
        elif dificultad == 2:
            return 12
        elif dificultad == 3:
            return 5
        else:
            print("Opción inválida. Elige un número entre 1 y 3. ")

#Funcion para guardar estadistica
def guardar_estadisticas(nombre, status, intentos):
    try:
        excel_document = openpyxl.load_workbook('juego.xlsx')
        hoja = excel_document['Hoja1']
    except FileNotFoundError:
        excel_document = openpyxl.Workbook()
        hoja = excel_document.active
        hoja.title = "Hoja1"
        hoja.append(["Nombre", "status", "Intentos"])  
        
    hoja.append([nombre, status, intentos])
    excel_document.save('juego.xlsx')
        
#Mostrar estadisticas
def mostrar_estadisticas():
    df = pd.read_excel('juego.xlsx', sheet_name='Hoja1')
    print(df.to_string(index=False))
    input("Presiona Enter para volver al menú principal...")



#Funcion Modo solitario
def modo_solitario():
        intento_maximo = dificultad()
    
        numero = random.randint(1, 1000)
        intento = 1
        
        while intento <= intento_maximo:
            guess_numero = int(input("Adivina el número entre 1 y 1000: "))
            
            # Comprobación de si el número está dentro del rango permitido
            while guess_numero < 1 or guess_numero > 1000:
                print("Número fuera de rango. Debes elegir un número entre 1 y 1000.")
                guess_numero = int(input("Adivina el número entre 1 y 1000: "))
            
            if guess_numero == numero:
                print("Correcto! Has ganado.")
                status = "Won" 
                nombre = input("Introduce tu nombre: ")
                guardar_estadisticas(nombre, status, intento)
                break
            elif guess_numero > numero:
                print("El número es menor." , "Intentos restantes:" + str(intento_maximo - intento))
            else:
                print("El número es mayor.",  "Intentos restantes:" + str(intento_maximo - intento))
            intento += 1
            
            
            if intento > intento_maximo:
                print("Game over.")
                status = "Lost"
                nombre = input("Introduce tu nombre: ")
                guardar_estadisticas(nombre, status, intento-1)    

#Funcion modo 2 jugadores 
def modo_dos_jugadores():
    intento_maximo = dificultad()
             
    numero = int(getpass("Jugador 1, introduce un número entre 1 y 1000: "))
    while numero < 1 or numero > 1000:
        print("Número inválido, introduce un número entre 1 y 1000.")
        numero = int(getpass("Jugador 1, introduce un número entre 1 y 1000: "))
        
    
    print("Jugador 2, adivina el número entre 1 y 1000: ")
    intento = 1

    while intento <= intento_maximo:
        guess_numero = int(input())

        # Comprobación de si el número está dentro del rango permitido
        while guess_numero < 1 or guess_numero > 1000:
            print("Número fuera de rango. Debes elegir un número entre 1 y 1000.")
            guess_numero = int(input())
            
        if guess_numero == numero:
            print("Correcto! Jugador 2 ha ganado.")
            status = "Won"
            nombre = input("Introduce el nombre del Jugador 2: ")
            guardar_estadisticas(nombre, status, intento)
            break
        elif guess_numero > numero:
            print("El número es menor.", "Intentos restantes:" + str(intento_maximo - intento))
            intento = intento + 1
        else:
            print("El número es mayor.", "Intentos restantes:" + str(intento_maximo - intento))
            intento = intento + 1
      
        if intento > intento_maximo:
            print("Game over.")
            status = "Lost"
            nombre = input("Introduce el nombre del Jugador 2: ")
            guardar_estadisticas(nombre, status, intento-1)
            

# Main game loop
while True:
    option = Menu()
    if option == 1:
        modo_solitario()
    elif option == 2:
        modo_dos_jugadores()
    elif option == 3:
        mostrar_estadisticas()
    elif option == 4:
        print("Adiós")
        break
    else:
        print("Opción inválida. Elige un número entre 1 y 4. \n")
```
